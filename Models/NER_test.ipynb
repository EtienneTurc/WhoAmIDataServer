```python
import pandas as pd
```

```python
import spacy
from spacy import displacy
from collections import Counter
import fr_core_news_md
nlp = fr_core_news_md.load()
```

```python
df = pd.read_json(r'data/gmail_turc_maj.txt', encoding = 'utf-8').body.to_frame()
```

```python
for el in df.loc[:30,'body']:
    print(displacy.render(nlp(str(el)), jupyter=True, style='ent'))
```

```python
sent = "Bonjour comment allez vous? Je m'appelle Paul et je suis un chien à Paris. Etes vous sur Londres? Moi non je suis chez google en datascience."
```

```python
print(displacy.render(nlp(sent), jupyter=True, style='ent'))
```
